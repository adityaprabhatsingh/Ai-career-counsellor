```python
##modeling a bank accoount

class BankAccount:
  def __init__ (self,balance,owner):
    self.balance=balance
    self.owner=owner

  def deposit(self,amount):
    self.balance+=amount
    print(f"{amount} is deposited. the new baknce is {self.balance} ")

  def withdraw(self,amount):
    if amount>self.balance:
      print("insufficient balance")


    else:
      self.balance-=amount
      print(f"{amount} is withdraw .the new balance is {self.balance}")

  def check_balance(self):
    return self.balance


acccount1=BankAccount(6000, "aditya")
acccount1.deposit(5000)

acccount1.withdraw(1000)
print(acccount1.check_balance())
```

```python
###inheritance in python

## parent class

class car:
  def __init__(self,window,door,enginetype):
    self.window=window
    self.door=door
    self.enginetype=enginetype


  def drive(self):
    print(f"this person drive a {self.enginetype} car")


car1=car(4,5,"petrol")
car1.drive()



```

```python
class tesla(car):
  def __init__ (self,window,door,enginetype,selfdrive):
    super().__init__(window,door,enginetype)
    self.selfdrive=selfdrive

  def selfdriving(self):
    print(f"this car support a self drive car {self.selfdrive}")
```

```python
tesla1=tesla(4,5,"electric",True )

tesla1.selfdriving()
```

```python
tesla1.drive()
```

```python
### multiple inhertance
##when a class inhertanvce a more than one base class

class animals:
  def __init__(self,name):
    self.name=name

  def speak(self):
    print("subclass 1")

class pet:
  def __init__(self,owner):
    self.owner=owner



##derived class

class dog(animals,pet):
  def __init__(self,name,owner):
    animals.__init__(self,name)
    pet.__init__(self,owner)

  def speak(self):
    return f"{self.name} say a woof"




###crete a object


dog1=dog("buddy","aditya")

print(dog1.speak())
print(f"owner is {dog1.owner}")
```

Polymorphism is an OOP concept where one method or object can perform different actions depending on the situation and it typically achieved with overriding and overloading

Method Overriding

Method overriding means:

Child class provides its own implementation of parent class method
Same method name
Same parameters

It is an example of runtime polymorphism.

Method Overloading

Method overloading means:

Same method name
Different parameters
In the same class

It is an example of compile-time polymorphism.

```python
#method overriding

# base class



class animals:
  def sound(self):
    return "sound of the animals"


##derived class
class dog(animals):
  def sound(self):
    return "woof"

#derived class 2

 class cat(animals):
  def sound(self):
    return "meow!"

dog1=dog()
print(dog1.sound())


cat1=cat()
print(cat1.sound())



```
